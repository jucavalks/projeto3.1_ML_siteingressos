```python
import pandas as pd

uri = "projects.csv"
dados = pd.read_csv(uri)
dados.head()
```

no nao finalizado: se é 1 = nao finalizado.. seria legal se fosse finalizado = sim (1). então veremos função map: o valor da esquerda da função vira o da direita:

```python
dados["finalizado"] = dados["nao_finalizado"].map({1: 0, 0: 1})
dados.head()
```

quero jogar fora as linhas com hora esperada = 0, pq nao faz mt sentido esse finalizados no hora esperada, n tem mt informação:

```python
dados = dados.query("horas_esperadas > 0")
dados.head()
```

```python
x = dados[["horas_esperadas", "preco"]]
y = dados["finalizado"]
```

```python
dados.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler


raw_x_treino, raw_x_teste, y_treino, y_teste = train_test_split(x, y, random_state = 42, test_size = 0.25, stratify = y)

scaler = StandardScaler()
scaler.fit(raw_x_treino) #vai aprender a reescalar os dados
#scaler preencha de acordo c os dados do treino

#agr treinei ent preciso reescalar
x_treino = scaler.transform(raw_x_treino)
x_teste = scaler.transform(raw_x_teste)
#fit() calcula os parâmetros de transformação no treino, e
#transform() aplica esses parâmetros tanto no treino quanto no teste.

modelo = SVC(gamma='auto')
modelo.fit(x_treino, y_treino)
previsoes = modelo.predict(x_teste)

acuracia = accuracy_score(y_teste, previsoes)
print(f"A acurácia foi de {acuracia*100: .2f}  %")
```

```python
previsoes_de_base = y_teste.sum() / len(y_teste)
print(f"A acurácia do modelo de base foi de {previsoes_de_base*100: .2f}  %")

# base: chuta q todo mundo vai finalizar, quantos realmente finalizaram
```

Imagine que temos dois pontos (A e B) e duas características:

Idade: (Varia de 1 a 100).

Renda Anual: (Varia de 10.000 a 1.000.000).

Para calcular a distância entre A e B, o algoritmo calcula o quadrado da diferença em cada dimensão: A distância final é $\sqrt{25 + 10.000}$.Note que a diferença na Renda (10.000) é 400 vezes maior que a diferença na Idade (25). No cálculo final:A Idade contribui com apenas 0.25% da distância total (25 de 10.025).A Renda contribui com 99.75% da distância total (10.000 de 10.025).Conclusão: "Peso Injusto"Neste caso, o algoritmo praticamente ignora a informação contida na variável Idade, pois a grande escala da Renda dominou completamente o cálculo. Ele conclui que os pontos A e B são diferentes quase que exclusivamente por causa da Renda, mesmo que a diferença de 5 anos na Idade possa ser, para o nosso problema, um fator muito mais relevante.

A Primeira Coluna (ex: 0.85593127, -1.70535306) representa a característica horas_esperadas após a padronização.

A Segunda Coluna (ex: 0.76211486, -0.59243057) representa a característica preco após a padronização.

Valores Próximos de Zero (ex: -0.23 ou 0.15): Indicam que aquele ponto de dados (aquela linha/projeto) tem valores de horas_esperadas ou preco que estão muito próximos da média do seu conjunto de treino.Valores Positivos Elevados (ex: 0.85): Indicam que o valor original (horas ou preço) está acima da média do seu conjunto de treino (cerca de $0.85$ desvios-padrão acima da média).Valores Negativos Elevados (ex: -1.70): Indicam que o valor original está abaixo da média do seu conjunto de treino (cerca de $1.70$ desvios-padrão abaixo da média).

A padronização (StandardScaler) é aplicada apenas à variável X (o raw_x_treino e raw_x_teste), pois são as entradas que precisam de escalas iguais.

A coluna finalizado (ou y_teste após a separação) não deve ser escalonada porque:

Já está na escala correta: O valor 0 significa uma coisa, e o valor 1 significa outra. Não há diferença de escala a ser corrigida.
